```python
import statsmodels.api as sm
```

```python
# define dependent variable
y_variable_name = 'DEATH_EVENT'
```

```python
# define alpha level
alpha_level = 0.05
```

```python
# to be used in the backward_stepwise_log_regression function it will help us computerize the process and decrease code
def logitregression(y, X):
    log_regress = sm.Logit(y,X)
    results_log = log_regress.fit()
    return results_log
```

```python
# purpose of this function is to execute a Backward stepwise selection with all columns that will break once all the predictor
# variables included are significant 
def backward_stepwise_log_regression(HFdata,y_variable_name ,alpha_level):
    X = HFdata.loc[:, HFdata.columns != y_variable_name] # Features
    X = sm.add_constant(X)
    y = HFdata[y_variable_name]
    while True:
        result = logitregression(y,X)
        pvalues = result.pvalues
        if max(pvalues)<= alpha_level:
            break
        for index,pvalue in pvalues.items():
            if pvalue == max(pvalues):
                X.drop(columns=[index], inplace = True)
    return result
    
```

```python
best_log_model = backward_stepwise_log_regression(HFdata,y_variable_name, alpha_level)
```

```python
best_log_model.summary()
```
